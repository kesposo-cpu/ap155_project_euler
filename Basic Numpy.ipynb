```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0, 5, 0.01)
freq1 = 1.0
freq2 = 10.0

y1 = np.sin(2*np.pi*freq1*t)
y2 = np.sin(2*np.pi*freq2*t)

product = y1 * y2

fig, ax = plt.subplots()

ax.plot(t, product, label="product")

ax.legend()
```

```python
t = np.linspace(0, 1, 100)

circle = np.sqrt(1 - t**2)

plt.plot(t, circle)
plt.axis('equal')
plt.title('Circle Plot')
plt.show()
```

```python
t = np.linspace(0, 1, 100)

circle = np.sqrt(1 - t**2)

fig, ax = plt.subplots()

count_in = 0
count_out = 0

for i in range(1000):
    x_i = np.random.random()
    y_i = np.random.random()
    if (x_i**2 + y_i ** 2 < 1):
        ax.scatter(x_i, y_i)
        count_in += 1
    else: 
        x_i**2 + y_i ** 2 > 1
        count_out += 1

print(count_in)
print(count_out)
print(4*(count_in/(count_in + count_out)))
plt.plot(t, circle)
```

```python
N = 1000000
x_i = np.random.random(size = N)
y_i = np.random.random(size = N)

is_inside = x_i**2 + y_i**2 < 1

print(is_inside)

fig, ax = plt.subplots()

ax.scatter(x_i[is_inside], y_i[is_inside])

4*np.sum(is_inside)/N
```
