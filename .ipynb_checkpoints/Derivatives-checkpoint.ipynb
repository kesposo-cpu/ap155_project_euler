```python
import numpy as np
import matplotlib.pyplot as plt

def f_func(x):
    return ((1 - np.cos(x)) / (x**2))

h = np.logspace(-1, -6, 6)
x = 0.004

def central_diff(x):
    h = np.logspace(-1, -6, 6)
    cent_diff = (f_func(x+h) - 2*f_func(x) + f_func(x-h)) / h**2
    analytic = (-4*(np.sin(x))/(x**3)) + ((-6*(np.cos(x)) + 6)/(x**4)) + ((np.cos(x))/(x**2))
    o = abs(cent_diff - analytic)
    return o
    
def plot(x):
    h = np.logspace(-1, -6, 6)
    o = central_diff(x)
    fig, ax = plt.subplots()
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(h,o)

plot(0.004)
```

```python
def new_func(x):
    return  (2*(np.sin(x/2))**2)/x**2

h = np.logspace(-1, -6, 100)
x = 0.004

def central_diff(x):
    h = np.logspace(-1, -6, 100)
    cent_diff = (new_func(x+h) - 2*new_func(x) + new_func(x-h)) / h**2
    analytic = (-4*(np.sin(x))/(x**3)) + ((-6*(np.cos(x)) + 6)/(x**4)) + ((np.cos(x))/(x**2))
    o = abs(cent_diff - analytic)
    return o
    
def plot(x):
    h = np.logspace(-1, -6, 100)
    o = central_diff(x)
    fig, ax = plt.subplots()
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(h,o)

plot(0.004)
```
